```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from pydantic import BaseModel
```

```python
# define state
class LoanState(TypedDict):

    applicant_name: str
    application_id: str
    credit_score: float
    risk_level: str
    approved: bool
    disbursed_amount: float






```

```python
# Step 1: Application
loan_state: LoanState = {
    "applicant": "Deepu Bandi",
    "application_id": "LN0303",
    "credit_score": 0.0,
    "risk_category": "",
    "is_approved": False,
    "disbursed_amount": 0.0

}


```

```python
def submit_application(state: LoanState):
    print("Step 1 - Application Submitted")
    return state

def credit_check(state: LoanState):
    state["credit_score"] = 720.0
    print("Step 2 - Credit Check:", state["credit_score"])
    return state

def risk_assessment(state: LoanState):
    score = state["credit_score"]
    if score >= 700:
        state["risk_level"] = "Low"
    elif score >= 600:
        state["risk_level"] = "Medium"
    else:
        state["risk_level"] = "High"
    print("Step 3 - Risk Assessment:", state["risk_level"])
    return state

def loan_approval(state: LoanState):
    state["approved"] = state["risk_level"] == "Low"
    print("Step 4 - Loan Approval:", state["approved"])
    return state

def disbursement(state: LoanState):
    if state["approved"]:
        state["disbursed_amount"] = 50000.0
    else:
        state["disbursed_amount"] = 0.0
    print("Step 5 - Disbursement:", state["disbursed_amount"])
    return state
```

```python
#define Graph and START/END
graph=StateGraph(LoanState)

#Adding Nodes to Graph
graph.add_node('submit_application',submit_application)
graph.add_node('credit_check',credit_check)
graph.add_node('risk_assessment',risk_assessment)
graph.add_node('loan_approval',loan_approval)
graph.add_node('disbursement',disbursement)


#Adding Edges to Graph

graph.add_edge(START,'submit_application')
graph.add_edge('submit_application','credit_check')
graph.add_edge('credit_check','risk_assessment')
graph.add_edge('risk_assessment','loan_approval')
graph.add_edge('loan_approval','disbursement')
graph.add_edge('disbursement',END)

#compile Workflow

workflow = graph.compile()
```

```python
#Executing Graph


initial_state: LoanState = {
    "applicant": "Deepu Bandi",
    "application_id": "LN0303",
    "credit_score": 0.0,
    "risk_category": "",
    "is_approved": False,
    "disbursed_amount": 0.0
}



final_state = workflow.invoke(initial_state)
print("Final State:", final_state)
```
